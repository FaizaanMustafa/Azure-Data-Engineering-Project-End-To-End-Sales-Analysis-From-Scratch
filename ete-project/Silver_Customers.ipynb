```python
from pyspark.sql.functions import *
from pyspark.sql.types import *
import time
```

```python
df = spark.read.format('parquet').load('abfss://sales@adls7428.dfs.core.windows.net/bronze/customers')
df.display()
```

```python
df = df.withColumn('domain',split(col('email'),'@')[1])
df.display()
```

```python
df.groupBy('domain').agg(count('customer_id').alias('total_cutomers')).sort(desc('total_cutomers')).display()
```

```python
df_gmail = df.filter(col('domain')=='gmail.com')
df_gmail.display()
time.sleep(5)

df_yahoo = df.filter(col('domain')=='yahoo.com')
df_yahoo.display()
time.sleep(5)

df_hotmail = df.filter(col('domain')=='hotmail.com')
df_hotmail.display()
time.sleep(5)


```

```python
df = df.withColumn('name',concat(col('first_name'),lit(' '),col('last_name')))
df = df.drop('first_name','last_name')
df.display()
```

```python
df.write.format('delta').mode('append').option('path','abfss://sales@adls7428.dfs.core.windows.net/silver/customers').saveAsTable('ete_project.silver.customers')
```
