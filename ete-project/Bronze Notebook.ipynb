```python
dbutils.widgets.text('folder_name','')
dbutils.widgets.text('source_sub_folder','')
dbutils.widgets.text('target_sub_folder','')
```

```python
var1 = dbutils.widgets.get('folder_name')
sub_folder = dbutils.widgets.get('source_sub_folder')
target_sub_folder = dbutils.widgets.get('target_sub_folder')
```

```python
df = spark.readStream.format('cloudFiles') \
                .option('cloudFiles.format', 'parquet') \
                .option('cloudFiles.schemaLocation', f'abfss://sales@adls7428.dfs.core.windows.net/{sub_folder}/checkpoint_{var1}') \
                .load(f'abfss://sales@adls7428.dfs.core.windows.net/{sub_folder}/{var1}')
df.display()
```

```python
df.writeStream.format('parquet').outputMode('append').option('checkpointLocation', f'abfss://sales@adls7428.dfs.core.windows.net/{target_sub_folder}/checkpoint_{var1}').option('path', f'abfss://sales@adls7428.dfs.core.windows.net/{target_sub_folder}/{var1}').trigger(once=True).start()
```
